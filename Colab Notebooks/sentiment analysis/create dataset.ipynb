```
import configparser
from textblob import TextBlob
import tweepy
import matplotlib.pyplot as plt
import regex as re
import spacy
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
#import googlemaps
import sys
import os
import pandas as pd
```

```
config=configparser.RawConfigParser()
config.read('/content/drive/My Drive/Colab Notebooks/uemprojectgroup covid_1.properties')
print(config.sections())
```

```
consumerKey=config.get('twitter','consumerKey')
consumerSecret=config.get('twitter','consumerSecret')
accessToken=config.get('twitter','accessToken')
accessTokenSecret=config.get('twitter','accessTokenSecret')
```

```
auth=tweepy.OAuthHandler(consumer_key=consumerKey,consumer_secret=consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)
```

```
import time
time.time()
t_end = time.time() + 60 * 1

covid_df=pd.DataFrame([])
api=tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)

while time.time() < t_end:
  print("\n\nIN WHILE\n")
  a_tweets=tweepy.Cursor(api.search, q="covid"+"-filter:retweets").items(100)
  covid_list=[]
  for i in a_tweets:
    covid_list.append([i.text,i.user.screen_name,i.user.location,i.geo])
  ncovid_df=pd.DataFrame(covid_list,columns=["tweets","user_name","location","geo"])

  print("length of ncovid_df:",len(ncovid_df))
  display(ncovid_df)

  covid_df=covid_df.append(ncovid_df,ignore_index=True)

  print("length of covid_df:",len(covid_df))
  display(covid_df)

  covid_df.drop_duplicates(subset=['tweets','user_name','location'],inplace=True,ignore_index=True)

  print("length of unique covid_df:",len(covid_df))
  display(covid_df)

print("\n\nOUT WHILE\n")

print(len(covid_df))
display(covid_df)

covid_df.to_csv('covid_csv.csv')
```
