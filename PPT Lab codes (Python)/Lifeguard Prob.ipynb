```python
import math
```

```python
x_l=int(input()) 
y_l=int(input()) 
x_w=int(input()) 
y_w=int(input()) 
f=float(input()) 
```

```python
print(x_l,",",y_l)
print(x_w,",",y_w)
print(f)
```

```python
def kindOfTotalTime(x_l,y_l,x_w,y_w,f,x):
    
    res= math.sqrt( (x_l-x)**2 + y_l**2 ) + f*math.sqrt( (x_w-x)**2 + y_w**2 )
    
    return res
```

```python
DX=0.001
if x_l == x_w:
    x="{:.6f}".format(x_l)
    print(x)
else:
    x=x_l
    factor=kindOfTotalTime(x_l,y_l,x_w,y_w,f,x)
    stepSize=1
    print("********************* step size =",stepSize,"*********************")
    count=0
    while True:
        count+=1
        x_prev=x
        factor_prev=factor
        x=x+stepSize
        factor=kindOfTotalTime(x_l,y_l,x_w,y_w,f,x)
        print("x prev=",x_prev," x=",x)
        print(factor_prev,factor)
        slope= (kindOfTotalTime(x_l,y_l,x_w,y_w,f,x+DX)-kindOfTotalTime(x_l,y_l,x_w,y_w,f,x))/DX
        
        stepSize=abs(slope) 
        print("********************* step size =",stepSize,"*********************")
    
        # if abs(x-x_prev)<0.000001:
        #     break
        if abs(x-x_prev)<0.000001 or x>x_w or factor_prev-factor<0: #or count>100:
            break
res="{:.6f}".format(x_prev)
print("output: (x,y)=(",res,",0)")
```
